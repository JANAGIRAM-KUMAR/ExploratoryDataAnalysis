```python
# Left Skewed Distribution
# With Using Distribution

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import math

# Dataset
X = np.array([30, 60, 70, 75, 80, 85, 90])

# Measures of Central Tendency
mean = np.mean(X)
median = np.median(X)
mode = stats.mode(X, keepdims=True).mode[0]

# Measures of Dispersion
variance = np.var(X)
std_dev = np.std(X)
skewness = stats.skew(X)

# Output
print("Left Skewed Distribution")
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Skewness:", skewness)

x = [i for i in range(0, 100)]
y = [math.exp(-0.05 * (100 - i)) * (100 - i) for i in x]

plt.figure()
plt.plot(x, y)
plt.title("Negative Skewed Distribution")
plt.xlabel("X")
plt.ylabel("Frequency")
plt.show()
```
