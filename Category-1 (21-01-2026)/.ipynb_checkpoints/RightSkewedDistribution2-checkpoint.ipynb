```python
# Right Skewed Distribution
# With Using Library

import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import mode, skew

X = np.array([10, 12, 13, 14, 15, 18, 40])

# Measures of Central Tendency
mean = np.mean(X)
median = np.median(X)
mode_value = mode(X, keepdims=True).mode[0]

# Measures of Dispersion
variance = np.var(X)
std_dev = np.std(X)
skewness = skew(X)

# Output
print("Right Skewed Distribution")
print("Mean =", mean)
print("Median =", median)
print("Mode =", mode_value)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
print("Skewness =", skewness)

x = [i for i in range(0, 100)]
y = [math.exp(-0.05 * i) * i for i in x]

plt.figure()
plt.plot(x, y)
plt.title("Positive Skewed Distribution")
plt.xlabel("X")
plt.ylabel("Frequency")
plt.show()
```
