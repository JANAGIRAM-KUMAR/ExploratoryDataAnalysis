```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
#Line Chart
plt.plot(iris["sepal_length"])
plt.title("Line Chart - Sepal Length")
plt.show()
```

```python
#Bar Chart
sns.barplot(x="species", y="sepal_length", data=iris)
plt.title("Bar Chart - Avg Sepal Length")
plt.show()
```

```python
#Histogram
plt.hist(iris["petal_length"], bins=10)
plt.title("Histogram - Petal Length")
plt.show()
```

```python
#Box Plot
sns.boxplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
#Violin Plot
sns.violinplot(x="species", y="petal_width", data=iris)
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
plt.show()
```

```python
#Bubble Chart
plt.scatter(iris["sepal_length"], iris["petal_length"],
            s=iris["petal_width"]*100, alpha=0.5)
plt.show()
```

```python
#Area Plot
iris[['sepal_length','sepal_width']].plot.area()
plt.show()
```

```python
#Stacked Plot
iris.groupby("species")[["sepal_length","sepal_width"]].mean().plot(kind="bar", stacked=True)
plt.show()
```

```python
#Pie Chart
iris["species"].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```

```python
#Table Chart
print(iris.describe())
```

```python
#Polar Chart
theta = np.linspace(0, 2*np.pi, len(iris))
r = iris["sepal_length"]

ax = plt.subplot(111, polar=True)
ax.plot(theta, r)
plt.show()
```

```python
#Lollipop Chart
x = iris["species"].value_counts().index
y = iris["species"].value_counts().values

plt.stem(x, y)
plt.show()
```

```python
#Heatmap
sns.heatmap(iris.corr(numeric_only=True), annot=True)
plt.show()
```

```python
#Pair Plot
sns.pairplot(iris, hue="species")
plt.show()
```
