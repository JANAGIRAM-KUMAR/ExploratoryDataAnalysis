```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#Line Chart
titanic["age"].plot()
plt.title("Age Line Chart")
plt.show()
```

```python
#Bar Chart
sns.barplot(x="class", y="survived", data=titanic)
plt.show()
```

```python
#Histogram
plt.hist(titanic["age"].dropna(), bins=15)
plt.show()
```

```python
#Box Plot
sns.boxplot(x="class", y="age", data=titanic)
plt.show()
```

```python
#Violin Plot
sns.violinplot(x="sex", y="age", data=titanic)
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x="age", y="fare", hue="survived", data=titanic)
plt.show()
```

```python
#Bubble Chart
plt.scatter(titanic["age"], titanic["fare"],
            s=titanic["fare"], alpha=0.3)
plt.show()
```

```python
#Area Plot
titanic[['age','fare']].dropna().plot.area()
plt.show()
```

```python
#Stacked Plot
# Create stacked data
stacked_data = (
    titanic
    .groupby(["class", "survived"], observed=False)
    .size()
    .unstack(fill_value=0)
)

# Plot stacked bar chart
stacked_data.plot(
    kind="bar",
    stacked=True
)

plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.title("Titanic Survival Status by Class (Stacked Bar Plot)")
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
#Pie Chart
titanic["sex"].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```

```python
#Table Chart
print(titanic.describe())
```

```python
#Polar Chart
# Remove rows where fare is null
clean_data = titanic["fare"].dropna()

# Create theta with same length
theta = np.linspace(0, 2*np.pi, len(clean_data))

r = clean_data.values

ax = plt.subplot(111, polar=True)
ax.plot(theta, r)

plt.title("Polar Plot of Titanic Fare")
plt.show()
```

```python
#Lollipop Plot
x = titanic["class"].value_counts().index
y = titanic["class"].value_counts().values

plt.stem(x, y)
plt.show()
```

```python
#Heatmap
sns.heatmap(titanic.corr(numeric_only=True), annot=True)
plt.show()
```

```python
#Pair Plot
sns.pairplot(titanic, hue="survived")
plt.show()
```
